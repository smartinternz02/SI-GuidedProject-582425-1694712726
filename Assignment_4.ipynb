```python
df2 = df.drop('quality',axis = 1)

for col in df2.columns:
  q1=df[col].quantile(0.25)
  q3=df[col].quantile(0.75)
  IQR=q3-q1
  upper_limit=q3+1.5*IQR
  lower_limit=q1-1.5*IQR
  df=df[(df[col]<upper_limit) & (df[col]>lower_limit)]
```

```python
for col in df.columns:
    sns.boxplot(x=df[col])
    sns.set(rc={'figure.figsize':(8,6)})
    plt.xlabel(col)
    plt.ylabel('Value')
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python

x = df.drop('quality',axis = 1)
x.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
```

```python
x_scaled= pd.DataFrame(scale.fit_transform(x),columns =x.columns)
x_scaled.head()
```

```python
y = df.quality
y.head()
```

```python
# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size = 0.2,random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# Model Training

from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
model2 =RandomForestClassifier(n_estimators=200,criterion='entropy')
model2.fit(x_train,y_train)
```

```python
r_y_predict = model2.predict(x_test)
r_y_predict_train = model2.predict(x_train)
```

```python
print('Testing Accuracy = ', accuracy_score(y_test,r_y_predict))
print('Training Accuracy = ', accuracy_score(y_train,r_y_predict_train))   # Hence on random forest we're getting accuracy 0.71
```
